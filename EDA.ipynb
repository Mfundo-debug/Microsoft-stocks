```python
# check for outliers
def box_plot(data):
    for i in data.columns:
        if data[i].dtype == 'float64' or data[i].dtype == 'int64':
            sns.boxplot(data[i])
            plt.show()
box_plot(data)
```

```python
# check for the number of outliers in data
def outliers(data):
    for i in data.columns:
        if data[i].dtype == 'float64' or data[i].dtype == 'int64':
            q1 = data[i].quantile(0.25)
            q3 = data[i].quantile(0.75)
            iqr = q3 - q1
            print("Number of outliers present in ",i,"column: ",data[(data[i] < (q1 - 1.5 * iqr)) | (data[i] > (q3 + 1.5 * iqr))][i].count())
outliers(data)
```

#### There are several ways to handle missing values in a dataset. The most common methods are:
* 1. Drop the rows or columns that contain missing values
* 2. Impute the missing values with a statistical measure such as mean, median, or mode
* 3. Impute the missing values with a constant value
* 4. Impute the missing values with a value estimated by another predictive model
* 5. Z-score Method: Calculate the z-score of each observation and remove observations that fall above a certain threshold
* 6. Interquartile Range Method: Calculate the IQR of each observation and remove observations that fall above a certain threshold
* 7. KNN Method: Use the KNN algorithm to predict missing values
* 8. MICE Method: Use the MICE algorithm to predict missing values
* 9. Linear Regression Method: Use linear regression to predict missing values
* 10. Multiple Imputation Method: Use multiple imputation to predict missing values
* 11. Domain Knowledge Method: Use domain knowledge to predict missing values
* 12. Robust Statistics: Use robust statistics to predict missing values 

```python
# Remove the outliers for each column
def remove_outliers(data):
    for i in data.columns:
        if data[i].dtype == 'float64' or data[i].dtype == 'int64':
            q1 = data[i].quantile(0.25)
            q3 = data[i].quantile(0.75)
            iqr = q3 - q1
            data = data[(data[i] > (q1 - 1.5 * iqr)) & (data[i] < (q3 + 1.5 * iqr))]
    return data
data = remove_outliers(data)
```

```python
# check for skewness excluding date column
def skewness(data):
    for i in data.columns:
        if data[i].dtype == 'float64' or data[i].dtype == 'int64':
            sns.distplot(data[i])
            plt.show()
skewness(data)
```

```python
#check for covariance
data.cov()
```

```python
# plot the covariance
sns.heatmap(data.cov(), annot = True)
plt.show()
```

```python
# Apply log transformation to reduce the skewness
data['Open'] = np.log(data['Open'])
data['High'] = np.log(data['High'])
data['Low'] = np.log(data['Low'])
data['Volume'] = np.log(data['Volume'])
data['Price'] = np.log(data['Price'])
```

```python
#data modelling
#split the data into input and output
x = data.drop(columns = ['Price','Date'])
y = data['Price']
#check the shape of input and output
print(x.shape, y.shape)
```

```python
#split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .25, random_state = 42)
#check the shape
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
#scale the data
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
#check the shape
print(x_train.shape, x_test.shape)
```

```python

# import the model library
from sklearn.linear_model import LinearRegression
# instantiate the model
lr = LinearRegression()
# fit the model
lr.fit(x_train, y_train)
# predict the model
y_pred = lr.predict(x_test)
# check the score
lr.score(x_train, y_train),lr.score(x_test, y_test)
```

```python
# import the library
from sklearn.metrics import mean_squared_error
# check the mean squared error
mean_squared_error(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import r2_score
# check the r2 score
r2_score(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import mean_absolute_error
# check the mean absolute error
mean_absolute_error(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import mean_squared_log_error
# check the mean squared log error
mean_squared_log_error(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import median_absolute_error
# check the median absolute error
median_absolute_error(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import explained_variance_score
# check the explained variance score
explained_variance_score(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import max_error
# check the max error
max_error(y_test, y_pred)
```

```python
# import the library
from sklearn.metrics import mean_poisson_deviance
# check the mean poisson deviance
mean_poisson_deviance(y_test, y_pred)
```

```python
#Apply moving average to the data
data['Open'] = data['Open'].rolling(12).mean()
data['High'] = data['High'].rolling(12).mean()
data['Low'] = data['Low'].rolling(12).mean()
data['Volume'] = data['Volume'].rolling(12).mean()
data['Price'] = data['Price'].rolling(12).mean()
#plot the moving average data
plt.plot(data['Open'], label = 'Open')
plt.plot(data['High'], label = 'High')
plt.plot(data['Low'], label = 'Low')
plt.plot(data['Volume'], label = 'Volume')
plt.plot(data['Price'], label = 'Price')
plt.legend()
plt.show()
```

```python
#Apply exponential moving average to the data
data['Open'] = data['Open'].ewm(12).mean()
data['High'] = data['High'].ewm(12).mean()
data['Low'] = data['Low'].ewm(12).mean()
data['Volume'] = data['Volume'].ewm(12).mean()
data['Price'] = data['Price'].ewm(12).mean()
#plot the exponential moving average data
plt.plot(data['Open'], label = 'Open')
plt.plot(data['High'], label = 'High')
plt.plot(data['Low'], label = 'Low')
plt.plot(data['Volume'], label = 'Volume')
plt.plot(data['Price'], label = 'Price')
plt.legend()
plt.show()
```

```python
#plot the moving average data with relation to x='Date and y='Price'
plt.plot(data['Date'], data['Price'], label = 'Price')
plt.legend()
plt.show()
```
